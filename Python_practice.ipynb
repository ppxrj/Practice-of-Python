```python
#Exercise 1
age=int(input("Enter your age: "))
if age>= 18:
  print("You are old enough to drive")
else:
  print("You need to wait", 18-age,"years")
```

```python
#Exercise 2
my_age=int(input("Enter my age: "))
your_age=int(input("Enter your age: "))
if my_age>your_age:
  print("I'm", my_age-your_age, "years older")
elif my_age<your_age:
  print("You're",your_age-my_age, "years older")
else:
  print("Same age")
```

```python
#Ex L2
score=int(input("Enter your score: "))
if score>=80:
  print("A")
elif 69<score<80:
  print("B")
elif 59<score<70:
  print("C")
elif 49<score<60:
  print("D")
else:
  print("F")
```

```python
#Ex L3
person={
  'first_name': 'Asabeneh',
  'last_name': 'Yetayeh',
  'age': 250,
  'country': 'Finland',
  'is_marred': True,
  'skills': ['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
  'address': {
      'street': 'Space street',
      'zipcode': '02210'
  }
}
skill_check=person.keys()
print(skill_check)
print(person.get('skills'))
print('Python' in person)
person = {'skills': ['JavaScript', 'React', 'Node']}

if all(skill in person['skills'] for skill in ['JavaScript', 'React']):
    print("Front end dev")
elif all(skill in person['skills'] for skill in ['Node', 'MongoDB', 'Python']):
    print("Back end dev")
elif all(skill in person['skills'] for skill in ['React', 'Node', 'MongoDB']):
    print("Full stack dev")
else:
    print("unknown title")
```

```python
count = 0
while count < 8:
    if count == 3:
        count = count + 1
        continue
    print(count)
    count = count + 1
```

```python
numbers = (0,1,2,3,4,5)
for number in numbers:
    print(number)
    if number == 3:
        continue
    print('Next number should be ', number + 1) if number != 5 else print("loop's end") # for short hand conditions need both if and else statements
print('outside the loop')
```
